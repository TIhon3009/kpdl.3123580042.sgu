```python
# ex1_calc.py
print("Calculator Basic")
# Nhập số a và b từ bàn phím
a = int(input("Mời bạn nhập số a: "))
b = int(input("Mời bạn nhập số b: "))
# Thực hiện phép cộng
c = a + b
# In kết quả
print("c = %d + %d = %d" % (a, b, c))
```

```python
# ex3_circle.py
import math
print("Tính chu vi và diện tích hình tròn")
# Nhập bán kính r (số thực)
r = float(input("Mời bạn nhập bán kính r: "))
# Tính chu vi và diện tích
chu_vi = 2 * math.pi * r
dien_tich = math.pi * r**2
# In kết quả
print("Chu vi = %.2f" % chu_vi)
print("Diện tích = %.2f" % dien_tich)
```
